# Listas

```python
numeros = [1,2,3,4]
numeros
```

```python
palabras = ["Hola", "como", "estas", "?"]
palabras
```

```python
datos = ["Un string", 156, -78.14, "Otro string", True]
datos
```

```python
nombres = ["Juan", "Facu", "Damian"]
nombres
```

## Las listas son muy parecidas a los strings ya que funcionan con índices.


```python
datos[0]
```

```python
datos[-1]
```

## También usan slicing como los strings

```python
datos[:3]
```

```python
datos[-2:]
```

## Otra cosa que permiten las listas es concatenar con otras listas

```python
numeros += [5, 6, 7, 8]
numeros
```

```python
datos+= [1, "Otro string mas", False]
datos
```

## Las listas son muy parecidos a los string, pero los string son inmutables, y las listas si! 

```python
pares = [0, 2, 4, 5, 8, 10]
pares
```

```python
pares[3] = 6
pares
```

# Funciones!  
## Algunas de las funciones que tienen las listas son para agregar items al final

```python
pares.append(12)
pares
```

```python
pares.append(7*2-5+18)
pares
```

```python
datos.append(True)
datos
```

##  Otra función que tienen las listas es eliminar el último item

```python
datos.pop()
datos
```

## Que pasa si quiero eliminar más de un item a la vez? 

```python
datos[:3] = []
datos
```

## Y si quiero eliminar todo el contenido?

```python
datos = []
datos
```

## Las listas tienen la capacidad de contener otras listas, a esto, se lo llama listas anidadas

```python
a = [1,2,3]
b = [4,5,6]
c = [7,8,9]
resultado = [a,b,c]
resultado
```

```python
resultado[1]
```

```python
resultado[1][0]
```
